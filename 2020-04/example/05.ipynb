```python
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns

rc = {
    'axes.unicode_minus': False,
    'font.family': 'Malgun Gothic'
}
sns.set_style('whitegrid', rc) # tick(각 그리드) 스타일 지정
sns.set_palette('bright') # 색상 팔렛트 지정

import pandas as pd
df = pd.read_excel('data/sample-melon.xlsx')
```

```python
print(df.shape)
df.head()
```

```python
df['가수'].value_counts().plot(kind='bar')
```

```python
plt.figure(figsize=(30, 8))
sns.countplot(data=df, x='가수')
plt.xticks(rotation=90)
```

```python
df['가수'].value_counts().sort_values(ascending=False)[:10]
```

```python
df.groupby('가수').size().sort_values(ascending=False)[:10]
```

```python
팔렛트 = sns.color_palette('husl')
series = df.groupby('가수').size()
series.plot(kind='bar', figsize=(30, 8), color=팔렛트)
```

```python
with sns.color_palette('husl') as 팔렛트:
    series = df.groupby('가수').size()
    series.plot(kind='bar', figsize=(30, 8), color=팔렛트)
```
